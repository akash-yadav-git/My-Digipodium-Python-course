```python
import pandas as pd
import numpy as np
import folium
```

```python
df_incidents=pd.read_csv('incidents.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df_incidents.info()
```

```python
df_incidents.dropna(inplace=True)
```

```python
df_incidents.head()
```

```python
lat,lng=37.77, -122.42 #san francisco
```

```python
map=folium.Map([lat,lng],zoom_start=11)
crimes=folium.map.FeatureGroup()
data=df_incidents[:1000]#limit
for lat,lng in zip(data.Y, data.X):
    crimes.add_child(
        folium.CircleMarker(
            [lat,lng],
            radius=3,
            color='green',
            fill=True,
            fill_color='blue',
            fill_opacity=.5
        )
    )
    map.add_child(crimes)
```
